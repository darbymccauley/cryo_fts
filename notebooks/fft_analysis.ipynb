```python
import numpy as np
from astropy import units as u
import matplotlib.pyplot as plt
```

```python
data = np.load('../data/trial-run-diode.npz')
RES = data['RES_mm'] * u.mm
NSTEPS = data['NSTEPS']
R = data['R_V'] * u.V
POS = data['MOTOR_POS_mm'] * u.mm
OPD = POS * 2

R_rebinned = R.reshape(NSTEPS, 50).mean(axis = 1)
power = np.abs(np.fft.fftshift(np.fft.fft(R_rebinned - np.mean(R_rebinned))))**2
wavenumbers = np.fft.fftshift(np.fft.fftfreq(int(NSTEPS), (RES * 2))).to(1/u.cm)
plt.figure()
plt.plot(wavenumbers, power)
plt.xlabel(f'Wavenumber {wavenumbers.unit}')
plt.ylabel(f'Power {power.unit}')
plt.xlim(0, wavenumbers.value.max())
plt.grid()
plt.show()
```
